```python
import matplotlib.pyplot as plt
import seaborn as sns

price_by_date = df.groupby('date')['price'].mean().reset_index()

plt.figure(figsize =(12,6))
sns.lineplot(data=price_by_date, x='date', y='price')
plt.title('average house price distribution by date ')
plt.xlabel('date')
plt.ylabel('average price')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

categorical_features_for_ploting = ['bedrooms','bathrooms','floors','waterfront','view','condition','grade']
numerical_features_for_ploting = ['sqft_living','sqft_above','sqft_living15']

for col in categorical_features_for_ploting:
    plt.figure(figsize =(10,6))
    sns.boxplot(data=df, x='col', y='price',hue=col, palette='viridis', legend=False)
    plt.title(f'normalize price distribution by {col}')
    plt.xlabel('col')
    plt.ylabel('normalized price')
    plt.grid(axis='y')
    plt.show()
    



for col in numerical_features_for_ploting:
    plt.figure(figsize =(10,6))
    sns.scatterplot(data=df, x='col', y='price',alpha=0.6)
    plt.title(f'normalize price vs  {col}')
    plt.xlabel('col')
    plt.ylabel('normalized price')
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset load
df = pd.read_csv("Exam_Score_Prediction.csv")

# Pastel blue background + overall style
plt.style.use("default")
sns.set_theme(context="talk", style="whitegrid")

PASTEL_BG = "#A7C7E7"   # background
CARD_BG   = "#FFFFFF"   # inner axes background

TEXT_DARK = "#0F172A"
TEXT_BODY = "#1F2933"

BLUE  = "#2563EB"
GREEN = "#16A34A"
ORANGE = "#F97316"
PINK   = "#EC4899"

def make_figure(figsize=(8,6), title=""):
    fig, ax = plt.subplots(figsize=figsize)
    fig.patch.set_facecolor(PASTEL_BG)     # outer background
    ax.set_facecolor(CARD_BG)              # chart card
    ax.spines[['top','right']].set_visible(False)
    ax.set_title(title, color=TEXT_DARK, weight="bold", pad=15)
    ax.tick_params(colors=TEXT_BODY)
    ax.yaxis.label.set_color(TEXT_BODY)
    ax.xaxis.label.set_color(TEXT_BODY)
    return fig, ax
```
